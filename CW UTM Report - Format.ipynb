```python
def campaign_name(campaign):
    if "AU-Cover-Wallet-Display-PMax" in campaign:
        return "AU-Cover-Wallet-Display-P-Max"
    elif "New-Display-Max-conv" in campaign:
        return "New-Display-P-Max-conv"
    elif "Cyber-P-max"in campaign:
        return "AU-Cyber-Display-P-Max"
    elif "CW-PMax"in campaign:
        return "AU-Cover-Wallet-Display-P-Max"
    elif "Landlord-Pmax"in campaign:
        return "Display-P-Max-Landlord"
    elif "Performance-Max"in campaign:
        return "New-Display-P-Max-conv"
    elif "P-Max"in campaign:
        return "New-Display-P-Max-conv"
    elif "Display-Max-Landlord"in campaign:
        return "Display-P-Max-Landlord"
    elif "New-Performance-Max"in campaign:
        return "New-Display-P-Max-conv"
    elif "Health-biz-pack-p-max"in campaign:
        return "New-Display-P-Max-conv"
    else:
        return "Other"
```

```python
def referrer_category(referrer_url):
    if 'business-insurance.aon.com.au' in referrer_url:
        return 'Business Insurance Website'
    elif 'https://www.aondirect.com.au/en-au/landlords-agents/' in referrer_url:
        return 'Landlord agents'
    elif 'https://www.aondirect.com.au/en-au/landlords' in referrer_url:
        return 'Landlords'
    elif 'https://www.aondirect.com.au/en-au/landlords/insurance/rental-protection-plus-advantage' in referrer_url:
        return 'Landlords Rental Protection Plus Advantage'
    elif 'https://www.coveriq.net/en-au/landlords/insurance/rental-protection-plus' in referrer_url:
        return 'Landlords Rental Protection Plus'
    elif 'https://www.aondirect.com.au/landing/insurance/professional-indemnity' in referrer_url:
        return 'Professional Indemnity aondirect'
    elif 'https://www.aondirect.com.au/landing/insurance/professional-indemnity-professions' in referrer_url:
        return 'Professional Indemnity professions'
    elif 'https://www.aondirect.com.au/landing/insurance/business-insurance-pack' in referrer_url:
        return 'Business Insurance Pack aondirect'
    elif 'https://www.aondirect.com.au/insurance/business-insurance-pack/health-business-insurance' in referrer_url:
        return 'Health Business Insurance Pack aondirect'
    elif 'Quotes.aondirect.com.au/magic-link' in referrer_url:
        return 'Magic link'
    elif '“aon.com.au”' in referrer_url:
        return 'Aon.com.au (from our National website)'
    elif 'https://www.aondirect.com.au/insurance/cyber' in referrer_url:
        return 'Cyber'
    elif 'https://www.aondirect.com.au/industry/real-estate/real-estate-agents' in referrer_url:
        return 'Real Estate Professional Indemnity aondirect'
    elif 'https://www.aondirect.com.au/landing/insurance/public-liability' in referrer_url:
        return 'Public Liability aondirect'
    elif 'https://www.aondirect.com.au/' in referrer_url:
        return 'Direct'
    elif 'https://www.google.com/' in referrer_url:
        return 'Google'
    else:
        return 'Other'
```

```python
df['referrer_category'] = df['referrer_url'].apply(referrer_category)
```

```python
def campaign_category(campaign):
    if "Real" in campaign:
        return "Real Estate"
    elif "IT-Consultant" in campaign:
        return "IT Consultant"
    elif "Management-Consultant" in campaign or "Management%20Consultants" in campaign:
        return "Management Consultant"
    elif "Design" in campaign:
        return "Design Professionals"
    elif "Business-Insurance" in campaign:
        return "Business Insurance"
    elif "Equine" in campaign:
        return "Equine"
    elif "P.I" in campaign:
        return "Professional-Indemnity"
    elif "Accountant" in campaign:
        return "Accountants"
    elif "Brand" in campaign:
        return "Brand"
    elif "Display" in campaign or "DA-" in campaign:
        return "Display"
    elif "Landlord" in campaign:
        return "Landlord"
    elif "Cyber-SME" in campaign:
        return "Cyber-SME"
    elif "Barristers" in campaign:
        return "Barristers"
    elif "Health" in campaign:
        return "Health"
    else:
        return "Others"
```

```python
df['campaign_category'] = df['campaign'].apply(campaign_category)
```

```python
def source_category(row):
    source = row['source']
    current_url = row['current_url']
    valid_sources = ["google", "google.com", "bing", "business-insurance.aon.com.au",
                     "linkedin", "facebook", "website", "biwebsite", "digital",
                     "eloqua", "newsletter", "easource1306", "customer.io",
                     "anca.org.au", "autopilot", "sf_au", "nutshell", 
                     "landlordsource", "bizpacksource", "pisource", "plsource",
                     "piprofsource", "sourcetestpl", "google.com.au", "sfmc", "undefined"]
    
    lower_source = source.lower()
    lower_current_url = current_url.lower()
    
    if lower_source in valid_sources or "_gl=" in lower_current_url:
        if lower_source in ["website", "biwebsite", "digital"] or "_gl=" in lower_current_url:
            return "website"
        else:
            return lower_source
    else:
        return "Others"
```

```python
df['source_category'] = df.apply(source_category, axis=1)
```

```python
def clean_currency(df,column_name):
    df[column_name] = df[column_name].str.replace('$','').str.replace(',','')
    df[column_name] = pd.to_numeric(df[column_name])
    return df
```

```python
df = clean_currency(df,column_name='premium_quoted')
df = clean_currency(df,column_name='total_revenue')
```
